```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis = np.arange(-100,100,0.1)
print(x_axis)
```

```python
mean = np.mean(x_axis)
std = np.std(x_axis)
print(mean,std)
```

```python
y_axis = 1/(std * np.sqrt(2 * np.pi)) * np.exp( - (x_axis - mean)**2 / (2 * std**2) )
```

```python
plt.plot(x_axis,y_axis,linewidth=3, color='r')
plt.show()
```
